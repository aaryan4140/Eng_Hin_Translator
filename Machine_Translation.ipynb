```python
#tokenize function

max_input_length = 128
max_target_length=128

source_lang= "en"
target_lang= "hi"

def process_function(sentence):
  input= [ex[source_lang] for ex in sentence["translation"]]
  target = [ex[target_lang] for ex in sentence["translation"]]


  #input tokenisation
  model_input = tokenizer(input, max_length= max_input_length, truncation= True)

  #output tokenisation
  with tokenizer.as_target_tokenizer():
    labels=tokenizer(target,max_length= max_target_length, truncation =True)

  model_input["labels"] = labels["input_ids"]
  return model_input
```

```python
tokenized_datasets = dataset.map(process_function, batched=True)
```

```python
tokenized_datasets
```

#Now its time to train the model through processed data

```python
model
```

```python
#loding pretrained model
model= TFAutoModelForSeq2SeqLM.from_pretrained(model)
```

```python
batch_size= 32
learning_rate= 0.001
weight_decay=0.01
num_train_epoch=1
```

```python
#we cant directly feed data into model
#so we are using data collator to combine token and model
data_collator = DataCollatorForSeq2Seq(tokenizer,model= model, return_tensors= "tf")
```

```python
train_dataset = model.prepare_tf_dataset(
    tokenized_datasets["test"], #taking test because system is not able to handle the train dataset
    batch_size= batch_size,
    shuffle= True,
    collate_fn = data_collator

)
```

```python
validation_dataset= model.prepare_tf_dataset(
    tokenized_datasets["validation"],
    batch_size= batch_size,
    shuffle= False,
    collate_fn = data_collator



)
```

```python
optimizer = AdamWeightDecay(learning_rate=learning_rate, weight_decay_rate=weight_decay)
```

```python
model.compile(optimizer=optimizer)
```

Finetuning model

```python
model.fit(train_dataset,validation_data= validation_dataset,epochs=2)
```

```python
model.save_pretrained("tf_model/")
```
